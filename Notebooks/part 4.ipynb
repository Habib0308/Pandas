```python
import pandas as pd
```

```python
pd.set_option('display.max_column', 150)
```

```python
url= '../input/stack-overflow-2018-developer-survey/survey_results_public.csv'
df= pd.read_csv(url)
```

```python
df.head(3)
```

# Apply Conditions to filter data

```python
df["Country"] == "Kenya"
```

```python
filt= (df["Country"] == "Kenya")
```

```python
df[filt]
```

# Using LOC for filtering data
just pass the filt variable inside loc method and you can also specify which columns you want to show data of.

```python
df.loc[filt, "CompanySize"]
```

# Using &, | for and or operations

```python
filt= (df["Country"] == "Kenya") & (df["DevType"] == "Full-stack developer")
```

```python
df.loc[filt]
```

# Negate Filter (Opposite of filter)
Use tilde sign ~

```python
df.loc[~filt]
```

# Filter multiple column values you want to see

```python
countries= ["India", "Spain, Pakistan", "Netherland", "Iran"]
filt= df["Country"].isin(countries)
```

```python
df[filt]
```

# Filter values of columns containing a specific character
Filter out all people with less than 2 years of experience and have UndergradMajors in Computer science.

It works like this that it says that the string within this column contains "Computer science"

[For more details](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.contains.html)

```python
filt = (df["UndergradMajor"].str.contains("Computer science", na= False))
df[filt]
```
