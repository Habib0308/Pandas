# Slicing rows using Indexers
loc

iloc

# iloc[]
Gets rows by index

```python
# Index Location
df.iloc[0]
```

You can get multiple rows by passing in list of indexes

```python
df.iloc[[0,1,70,5]]
```

Access rows and columns both

```python
df.iloc[[0,56,64], [0,5,6]]
```

# loc[]
Gets rows by labels of columns

```python
df.loc[[0,56,64], ['FormalEducation','Country']]
```

Get all columns from a range of certain column to a column

```python
df.loc[[1,2,3,4], 'Hobby':'DevType']
```
